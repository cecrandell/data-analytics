```python
import numpy as np
import matplotlib.pyplot as plt

x_data = np.arange(1,13,1)
fahrenheit_points = [39,42,51,62,72,82,86,84,77,65,55,44]

plt.plot(x_data,fahrenheit_points)
plt.show()
```

```python
celsius_points = [(x-32)*0.56 for x in fahrenheit_points]
plt.plot(x_data,celsius_points)
plt.show()
```

```python
plt.plot(x_data,fahrenheit_points)
plt.plot(x_data,celsius_points)
plt.show()
```
